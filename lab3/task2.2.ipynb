```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

получение набора данных Iris

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
print(df)
```

обучение модели

```python
clf = LogisticRegression(random_state=0, max_iter=1000)
clf.fit(df.drop(["target"], axis=1), df["target"])
y_pred = clf.predict(df.drop(["target"], axis=1))
```

отрисовка полученных даных

```python
colors = ['red', 'green', 'blue']
plt.figure(figsize=(12, 5))
for i, name in enumerate(iris.target_names):
    first = df[y_pred == i]["petal length (cm)"]
    second = df[y_pred == i]["petal width (cm)"]
    plt.scatter(first, second, color=colors[i],label=name)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.legend()
```

сравнивая с рисунком, полученным в лабораторной 3.1, можно сделать вывод, что они похожи. различаия есть только у нескольких точек virginica и versocolor
